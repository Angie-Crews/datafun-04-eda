```python
import numpy as np

import matplotlib.pyplot as plt

# Draw a fun shape: a colorful star
theta = np.linspace(0, 2 * np.pi, 11)
radii = np.array([1, 0.4] * 5 + [1])
x = np.cos(theta) * radii
y = np.sin(theta) * radii

plt.figure(figsize=(5, 5))
plt.fill(x, y, color='#e67e22', alpha=0.7, edgecolor='#9b59b6', linewidth=3)
plt.title("Fun Star Shape!")
plt.axis('equal')
plt.axis('off')
plt.show()
```

```python
# Create a pandas DataFrame from the x and y arrays
star_points = pd.DataFrame({'x': x, 'y': y})
print(star_points.head())
```
